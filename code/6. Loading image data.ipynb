## Loading images using torchvision image folder
> image files should be located in a specific way: root/class/image.png

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt
import torch
from torchvision import datasets, transforms
import helper1
```

```python
file_path_train = '../data/cat_dog/train'
```

#### load data

```python
trans = transforms.Compose([transforms.Resize(255), # resize images to 255 by 255 pixels
                  transforms.CenterCrop(224), # crop the center
                  transforms.ToTensor()]) # convert them to tensors
```

```python
dataset = datasets.ImageFolder(file_path_train,transform=trans) # load sets using ImageFolder
```

```python
dataloader = torch.utils.data.DataLoader(dataset, batch_size = 32, shuffle = True) # generator
```

```python
# change this to the trainloader or testloader 
data_iter = iter(dataloader)

images, labels = next(data_iter)
fig, axes = plt.subplots(figsize=(10,4), ncols=4)
for ii in range(4):
    ax = axes[ii]
    helper1.imshow(images[ii], ax=ax, normalize=False)
```

#### Load data using different transforms for test and train

```python
data_path = '../data/cat_dog/'
train_transform = transforms.Compose([transforms.RandomRotation(30),
                                      transforms.RandomResizedCrop(224),
                                      transforms.RandomHorizontalFlip(),
                                      transforms.ToTensor()]) # convert them to tensors
test_transform = transforms.Compose([transforms.Resize(255),
                                    transforms.CenterCrop(224),
                                    transforms.ToTensor()
    
])

train_set = datasets.ImageFolder(data_path+'train',transform=train_transform) # load sets using ImageFolder
test_set = datasets.ImageFolder(data_path+'test',transform=test_transform) # load sets using ImageFolder
```

```python
trainloader = torch.utils.data.DataLoader(train_set, batch_size = 32) # generator
testloader = torch.utils.data.DataLoader(test_set, batch_size = 32) # generator
```

```python
# change this to the trainloader or testloader 
data_iter = iter(trainloader)

images, labels = next(data_iter)
fig, axes = plt.subplots(figsize=(10,4), ncols=4)
for ii in range(4):
    ax = axes[ii]
    helper1.imshow(images[ii], ax=ax, normalize=False)
```

```python
# change this to the trainloader or testloader 
data_iter = iter(testloader)

images, labels = next(data_iter)
fig, axes = plt.subplots(figsize=(10,4), ncols=4)
for ii in range(4):
    ax = axes[ii]
    helper1.imshow(images[ii], ax=ax, normalize=False)
```
